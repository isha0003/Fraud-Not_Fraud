```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df=pd.read_csv('/Users/ishashree/Downloads/creditcard.csv')
df.sample(20)
```

```python
df.describe()
```

```python
sns.scatterplot(x="Amount",y="Class",data=df)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rc=RobustScaler()
df_s=df[["Time","Amount"]]
df_s.sample(10)
```

```python
df
```

```python
df_s=pd.DataFrame(rc.fit_transform(df_s))
df_s.columns=["scaled_time","scaled_amount"]
df=pd.concat([df,df_s],axis="columns")
df.drop(["Time","Amount"],axis="columns",inplace=True)
df.head()
```

```python
df["Class"].value_counts()
```

```python
fraud=df[df["Class"]==1]
non_fraud=df[df["Class"]==0]
#non_fraud=non_fraud.sample(frac=1)
non_fraud=non_fraud[:492]
new=pd.concat([non_fraud,fraud])
new=new.sample(frac=1)   #(frac=1) randomizes the data
new
```

```python
new["Class"].value_counts()
```

```python
new
```

```python
X=new.drop(["Class"],axis="columns")
y=new["Class"]
```

```python
from sklearn.linear_model import LogisticRegression
lr=  LogisticRegression()
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr.fit(X_train,y_train)
```

```python
y_predicted=lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
dd= confusion_matrix(y_test,y_predicted)
dd
```

```python
print(classification_report(y_test,y_predicted))
print('\n\n')
print(confusion_matrix(y_test,y_predicted))
print('\n')
print('accuracy is --> ',round(accuracy_score(y_test,y_predicted),2))
```
